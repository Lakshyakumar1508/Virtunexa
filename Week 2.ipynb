```python
import os
import shutil

# Define the directory to organize"
directory = r"C:\Users\laksh\OneDrive\Pictures\Screenshots"

# Create a mapping of extensions to folder names
EXTENSION_FOLDERS = {
    "Images": [".jpg", ".jpeg", ".png", ".gif", ".bmp"],
    "Documents": [".pdf", ".docx", ".txt", ".xlsx", ".pptx"],
    "Videos": [".mp4", ".mkv", ".avi", ".mov"],
    "Audio": [".mp3", ".wav", ".aac"],
    "Archives": [".zip", ".rar", ".tar", ".7z"],
    "Code": [".py", ".js", ".html", ".css", ".java", ".cpp"]
}

def organize_files(directory):
    if not os.path.exists(directory):
        print("Directory does not exist!")
        return

    for file in os.listdir(directory):
        file_path = os.path.join(directory, file)

        if os.path.isfile(file_path):
            file_ext = os.path.splitext(file)[-1].lower()

            # Find the appropriate folder
            folder_name = "Others"
            for category, extensions in EXTENSION_FOLDERS.items():
                if file_ext in extensions:
                    folder_name = category
                    break

            # Create the folder if it doesnâ€™t exist
            folder_path = os.path.join(directory, folder_name)
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)

            # Move the file
            shutil.move(file_path, os.path.join(folder_path, file))
            print(f"Moved: {file} -> {folder_name}/")

if __name__ == "__main__":
    organize_files(directory)
```
